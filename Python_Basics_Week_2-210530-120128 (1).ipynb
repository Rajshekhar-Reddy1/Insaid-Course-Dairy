#### **1.2 While Loop in Python**

##### **1.2.1 while loop**

Use **while loop** to iterate over a block of code as long as the **test expression** (also called test condition) is **true**.

- The body of the loop is entered only if the test_expression evaluates to True. 
- This process continues until the test_expression evaluates to False.


**Flowchart of <u>while</u> loop**

<img height="300" width="300" src="https://github.com/suchitmajumdar/LearnPython/blob/master/images/while_0.png?raw=true"></img>


**Syntax -**

    while test_expression:
        
        Body of while


**Example -**

````
a = 5

while (a<10):
  print (a)
  #a+=1
print('outside while loop')
````

````
len(myList)
````

````
# Find the sum of all numbers in a list

myList = [1, 2, 3, 4, 5]

sum = 0
index = 0

while index < len(myList):
    sum += myList[index]                      # sum = sum + myList[index]
    index += 1                                # index = index + 1
print('Sum of the array is: {}'.format(sum))
````

#### **1.3 For Loop**

<b>for</b> loop is a looping technque to **iterate over a sequence (list, tuple, string)**/ or other iterable objects. 

- Here, element is the variable that takes the value of the item inside the sequence on each iteration.

- Loop continues until we reach the last item in the sequence. 

**Flowchart of <u>for</u> loop**

<img height="300" width="300" src="https://github.com/suchitmajumdar/LearnPython/blob/master/images/for_0.png?raw=true"></img>


**Syntax -**

    for element in sequence :

        'for' code block  
        

**Example -**

````
# Sum of all numbers in a list

myList = [1, 2, 3, 4, 5]

sum = 0
#iterating over the list
for element in myList:
    sum += element
    
print("Sum of the aray is: {}".format(sum))
````

- **range(start,stop,step size)**

````
range(5)
````

````
# Print range of 5
for a in range(1,5,2):
    print(a)
````

- The **break** and **continue** statements are used in these cases.

#### Examples of break and continue

````
# Use of break
names = ['Suchit', 'Rakesh', 'Roshni']
for i in names:               # Let us iterate over the names list
    if i == 'Rakesh':
        break
    print(i)
else:
    print('For is completed and we are in else part')
print("Totally outside the for & else loop")
````

````
# Use of continue
names = ['Suchit', 'Rakesh', 'Roshni']

for name in names:               # Let us iterate over the names list
    if name == 'Rakesh':
        continue
    print(name)                  # If continue condition is satisfied we skip this line and carry on with the next iteration
   else:
    print('For is completed and we are in else part')
print("Totally outside the for & else loop")
````

#### Example: First 3 divisors of a number except 2 & 3

````
inputNumber = input("Please enter a positive number: ")
number = int(inputNumber)
listOfDivisors = [1]
for divisor in range(2,int(number/2)+1):
    if number%divisor == 0:
        if divisor in [2,3]:
            continue
        elif len(listOfDivisors) == 3:
            break
        listOfDivisors.append(divisor)
print(listOfDivisors)
````

#### **2.1 Python List**

- **List** is an **ordered sequence of items**. It is one of the most used datatype in Python and is very flexible. 

- All the items in a list **do not need** to be of the **same type.**

- Declaring a list is pretty straight forward. **Items separated by commas** are enclosed within square brackets **[ ]**.

````
myList = [1, 2, 3.3, "ThirdElement","More values"]
print(myList)  
type(myList)      
````

````
#print element through index position (it begins from 1)
print(myList[4])
````

- **Lists are mutable**, which means, value of elements of a list can be changed.

````
myList[3] = "Pankaj"
print(myList)
````

````
#remove a particular item
myList.remove('Pankaj')
print(myList)
````

````
#remove item from a particular index
myList.pop(2)
print(myList)
````

````
myList.insert(2,'a')
print(myList)
````

````
#add item at last index 
myList.append("LastElement")
print(myList)
````

#### **2.2 Python Tuple**

- **Tuple** is an **ordered sequence** of items same as list.
- The only difference is that **tuples are immutable**. Tuples once created cannot be modified.
- Tuples are used to write-protect data and are usually **faster than list** as it cannot change dynamically.
- It is defined within parentheses <b>()</b> where items are separated by commas.

````
my_tuple = (1, 2, 3.3, "ThirdElement", "FourthElement")
print(my_tuple)
````

````
#read element by corresponding index
print(my_tuple[4])
````

#### **2.3 Sets**

- Set is an **unordered collection of unique items**.
- Set is defined by values separated by comma inside curly braces **{}**.

````
my_Set = {10, 20, 30, 40, 50}
print(my_Set)
````

````
my_Set = {10, 20, 30, 30, 40,40,40,40}    # only uniques values are considered, duplicates will be removed
print(my_Set)
````

````
print(type(my_Set))
````

````
# cannot print a particular element as items are unordered
print(my_Set[2])
````

#### **2.4 Dictionary**

- **Dictionary** is an **unordered collection of key-value pairs**.
- It is generally used when we have a huge amount of data.
- We must know the **key** to **retrieve the value**.
- In Python, dictionaries are defined within curly braces {} with each item being a pair in the form **key:value**.
- Key and value can be of any type.

````
my_Dict = {'key1': "value1", 'key2': "value2", 'key3': "value3"}
print(my_Dict['key1'])
````

````
# update a value 
my_Dict['key2'] = "value2.1"
print(my_Dict)
````

````
# add a new key and value
my_Dict["key4"] = "value4"
print(my_Dict)
````

````
# delete dictionary elements
del my_Dict["key4"]
print(my_Dict)
````

# **3. Functions**
---

A group of related statements that perform a specific task is called as a <b>Function</b>.

**Syntax -**


    def function_name(arguments):
    
        """
        Doc String
        """
    
        Statement(s)

- **def** keyword notifies the **start of function header**

- **Arguments** (parameters) through which we **pass values** to a function. These are *optional*

- A **colon(:)** to mark the **end of funciton header**

- **Doc string** describe **what the function does**. This is optional

- **return** statement to **return a value from the function**. This is optional

**Example -**

````
def sum(numbers):
    
    a = 0
    
    for number in numbers:
        a += number
    return a
    
````

````
sum([1,2,3,4,5,7,8,9,10,11,23,13,4,5,5])
````

````
sum([1,2,3])
````

````
a,b,c,d = 1,2,3,4
sum = a + b + c + d
sum
````

````
a,b,c,d,e,f = 1,2,3,4,5,6
sum = a + b + c + d + e + f
sum
````

**Function Call -**

Once we have defined a function, we can call it from anywhere

````
sumOfNumbers = sum([2, 4, 6])

print(sumOfNumbers)
````

````
sumof4 = sum([1,2,3,4])
print(sumof4)
````
